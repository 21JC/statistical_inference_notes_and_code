```python
import seaborn as sns
import math
import numpy as np
import matplotlib.pyplot as plt
import random
def binomial_prob(theta,x,n):
    
    return (math.factorial(n)/(math.factorial(x)*math.factorial(n-x)))*(theta**x)*(1-theta)**(n-x)
```

```python
binomial_prob(0.5,8,10)
```

```python
idx = []
prob = []
for i in range(11):
    idx.append(i)
    prob.append(binomial_prob(0.5,i,10))
```

```python
sns.pointplot(x=idx,y=prob)
```

```python
# Plot likelihood
# Let's say we got 8 heads from 10 flips

idx = []
likelihood = []
for i in np.arange(0,1.1,0.1):
    idx.append(i)
    likelihood.append(binomial_prob(i,8,10))
```

```python
sns.pointplot(x=idx,y=likelihood)
```

```python
# Plot likelihood
# Let's say we got 0 heads from 10 flips

idx = []
likelihood = []
for i in np.arange(0,1.1,0.1):
    idx.append(i)
    likelihood.append(binomial_prob(i,0,10))
```

```python
sns.pointplot(x=[round(x,2) for x in idx],y=likelihood)
plt.xticks(rotation=70)
plt.show()
```

```python
from scipy.stats import beta
```

```python
# New born
a, b = 11,11

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)

plt.figure(figsize=(7,7))
plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a, b), 'r-')
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

```python
# true believer

a, b = 1,0.5

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)

plt.figure(figsize=(7,7))
plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a, b), 'r-',label="prior")
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')


# new a = 1+10 = 11
# new beta = 0.5+20-10 = 10.5
a, b = 11,10.5

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)

plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a, b), 'b-',label="posterior")
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.legend(loc="upper left")
plt.show()
```

```python
# strong skeptical

a, b = 100,100

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)

plt.figure(figsize=(7,7))
plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a, b), 'r-',label="prior")
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')


# new a = 100+50 = 150
# new beta = 100+100-50 = 150
a, b = 150,150

x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)

plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a, b), 'b-',label="posterior")
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.legend(loc="upper left")
plt.show()
```
