```python
import numpy as np
import random
import seaborn as sns
```

```python
num_sim = 100000
sample_size = 26

# simulated group's IQ with mean of 106
sim_sd = 15
sim_mean = 106
sim = np.random.normal(loc=sim_mean, scale=sim_sd, size=sample_size)

# normal group's IQ with mean of 100
norm = np.random.normal(loc=100, scale=sim_sd, size=sample_size)
```

```python
sns.distplot(sim)
sns.distplot(norm)
```

```python
print(sim.mean())
print(norm.mean())
```

```python
# Run 100 times, plot the distribution of P-value
```

```python
# Equal sample size & variance
import scipy.stats
def t_test_equal_var(x1,x2,sample_size):
    t = (x1.mean() - x2.mean())/ \
        ((x1.std()**2 + x2.std()**2)/sample_size )**0.5 
    # two-sided
    p_val = scipy.stats.t.sf(abs(t), df=sample_size)*2
    return t, p_val
```

```python
t_test_equal_var(sim,norm,sample_size)
```

```python
from scipy import stats
stats.ttest_ind(sim,norm,equal_var=True)

# Results slightly differ, possibly due to different implementation of Bessel's correction
```

```python
p_val = []
# Let's actually use one-sample t-test to be consistent with the tutorial
for i in range(num_sim):
    # simulated group's IQ with mean of 106
    sim = np.random.normal(loc=sim_mean, scale=sim_sd, size=sample_size)
    p_val.append(stats.ttest_1samp(sim, popmean=100).pvalue)
    if i % 5000 == 0:
        print(f"finished {i}th simulation")
```

```python
sns.histplot(p_val,bins=20)
```
