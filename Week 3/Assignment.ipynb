```python
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

n = 200
D = 0.0
SD = 1

p = np.zeros(n)
x = np.zeros(n + 10)
y = np.zeros(n + 10)

for i in range(10, n + 10):
    x[i] = np.random.normal(0, SD)
    y[i] = np.random.normal(D, SD)
    z = ttest_ind(x[:i+1], y[:i+1], equal_var=True)
    p[i - 10] = z.pvalue


plt.plot(np.arange(10, n + 10), p, linewidth=2)
plt.axhline(0.05, color="darkgrey", linestyle="--", linewidth=2)
plt.xlabel("sample size")
plt.ylabel("p-value")
plt.xticks(np.linspace(10, n + 10, 5))
plt.ylim(0, 1)
plt.xlim(10, n + 10)
plt.show()

print("The lowest p-value was observed at sample size", np.argmin(p) + 10)
print("The p-value dropped below 0.05 for the first time as sample size", np.where(p < 0.05)[0][0] + 10)
```

# Optional Stopping

```python
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

N = 100
Looks = 5
nSim = 50000
alpha = 0.05
D = 0

LookN = np.ceil(np.linspace(0, N, Looks + 1))
LookN = LookN[1:]
LookN = LookN[LookN > 2]
Looks = len(LookN)

matp = np.empty((nSim, Looks))
matp[:] = np.nan

SigSeq = np.zeros(Looks)
OptStop = np.zeros(nSim)
p = np.zeros(nSim)

for i in range(nSim):
    x = np.random.normal(0, 1, N)
    y = np.random.normal(D, 1, N)
    
    for j in range(Looks):
        z = ttest_ind(x[:int(LookN[j])], y[:int(LookN[j])], equal_var=True)
        # p-value for each sim
        matp[i, j] = z.pvalue
    if i % 10000 == 0:
        print(f"Loop {i + 1} of {nSim}")

# For each look, what is the total # of sig results
for i in range(Looks):
    SigSeq[i] = np.sum(matp[:, i] < alpha)

    
# For each sim, if we have a sig result, record the first time, else, record the last time (looks - 1)
for i in range(nSim):
    below_alpha = np.where(matp[i, :] < alpha)
    if below_alpha[0].size > 0:
        OptStop[i] = np.min(below_alpha)
    else:
        OptStop[i] = Looks - 1
    
    

for i in range(nSim):
    p[i] = matp[i, int(OptStop[i])]

breaks = 100
plt.hist(p, bins=breaks, color="grey", edgecolor="black")
plt.axhline(nSim / breaks, color="red", linestyle="--")
plt.show()

print(f"Type 1 error rates for look 1 to {Looks}: respectively {SigSeq / nSim}")
print(f"Type 1 error rate when only the lowest p-value for all looks is reported: {np.sum(p < alpha) / nSim}")
```
